### Criação do Tipo

```python
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
```

### Criação da Toolbox

```python
# Estou criando essa função porque quero que cada caracteristica de um indivíduo tenha apenas 5% de chance de ser 1.
# Assim, cada indivíduo nascerá bem "fraco" e será possível acompanhar as gerações evoluindo mais gradativamente.

def randomBit():
    a = randint(0,99)
    if a < 5:
        return 1
    else:
        return 0
```

```python
toolbox = base.Toolbox()

toolbox.register("attr_bool", randomBit)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=10)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
bit = toolbox.attr_bool()
ind = toolbox.individual()
pop = toolbox.population(n=3)

print("bit é do tipo %s e seu valor é: %s\n" % (type(bit), bit))
print("ind é do tipo %s e contém %d bits:\n%s\n" % (type(ind), len(ind), ind))
print("pop é do tipo %s e contém %d individuos:\n%s" % (type(pop), len(pop), pop))
```

### Função de Avaliação

```python
def evalOneMax(individual):
    return (sum(individual), )
```

### Operadores Genéticos

```python
toolbox.register("evaluate", evalOneMax)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.10)
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
indAntes = toolbox.individual()
print("Indivíduo:\t\t\t", indAntes)

indDepois = toolbox.clone(indAntes)
toolbox.mutate(indDepois)
print("Individuo após a função mutate:\t", indDepois)

if indAntes == indDepois:
    print("\nNão houve mutação.")
else:
    print("\nHouve mutação.")
```

```python
mutant = toolbox.clone(ind)
print(mutant is ind)
print(mutant == ind)
```

### Evoluindo a População

```python
# Para rodar o código com número de gerações e indivíduos diferentes dos sugeridos:
nGeracoes = 50
nIndividuos = 25
```

```python
def main():
    pop = toolbox.population(n=nIndividuos)
    hof = tools.HallOfFame(1)
    stats = tools.Statistics(lambda ind: ind.fitness.values)
    
    stats.register("avg", np.mean)
    stats.register("min", np.amin)
    stats.register("max", np.amax)
    
    pop, logbook = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2, ngen=nGeracoes, stats=stats, halloffame=hof, verbose=True)
    
    return pop, logbook, hof
```

```python
if __name__ == "__main__":
    pop, log, hof = main()
    print("\nO melhor indivíduo é: %s\nCom fitness: %s" % (hof[0], hof[0].fitness))
    
    gen, avg, min_, max_ = log.select("gen", "avg", "min", "max")
    plt.plot(gen, avg, label="Média", linewidth=3)
    plt.plot(gen, min_, label="Mínimo", linewidth=3)
    plt.plot(gen, max_, label="Máximo", linewidth=3)
    plt.xlabel("Generation")
    plt.ylabel("Fitness")
    plt.xlim(0, nGeracoes)
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()
```
