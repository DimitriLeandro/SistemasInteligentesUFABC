### Importação e preparação dos dados

```python
from sklearn.datasets import load_digits

dataset = load_digits()
X = dataset.images
y = dataset.target

X = X.reshape(X.shape[0], X.shape[1] * X.shape[2])

#verificando se esta ok
X.shape
```

#### Divisão dos dados em grupos para treinamento e teste.

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.25)
```

### Linear Discriminant Analysis (LDA)

#### Treinamento

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

LDA = LinearDiscriminantAnalysis().fit(X_train, y_train)
```

#### Classificação e Resultados

```python
y_pred_LDA = LDA.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("Acurácia: \n", accuracy_score(y_test, y_pred_LDA), "\n")
print("Matriz de Confusão: \n", confusion_matrix(y_test, y_pred_LDA))
```

### Support Vector Machine (SVM)

#### Treinamento

```python
from sklearn.svm import LinearSVC

SVM = LinearSVC().fit(X_train, y_train)
```

#### Classificação e Resultados

```python
y_pred_SVM = SVM.predict(X_test)

print("Acurácia: \n", accuracy_score(y_test, y_pred_SVM))
print("\nMatriz de Confusão: \n", confusion_matrix(y_test, y_pred_SVM))
```

### Artificial Neural Networks (ANN)

#### Treinamento

```python
from sklearn.neural_network import MLPClassifier

ANN = MLPClassifier(hidden_layer_sizes=(50)).fit(X_train, y_train)
```

#### Classificação e Resultados

```python
y_pred_ANN = ANN.predict(X_test)

print("Acurácia: \n", accuracy_score(y_test, y_pred_ANN))
print("\nMatriz de Confusão: \n", confusion_matrix(y_test, y_pred_ANN))
```

#### Análise para diferentes quantidades de neurônios

```python
vetorAcuracias = []

for qtdNeuronios in range(1, 51):
    ANN = MLPClassifier(hidden_layer_sizes=(qtdNeuronios)).fit(X_train, y_train)
    y_pred_ANN = ANN.predict(X_test)
    vetorAcuracias.append(accuracy_score(y_test, y_pred_ANN))
    print("Acurácia usando", qtdNeuronios, "neurônio(s):\t", "{0:.5f}".format(vetorAcuracias[qtdNeuronios - 1]))
```

```python
from numpy import average, amin, amax, median, var, std
from scipy import stats

print("Mínima:\t\t", "{0:.5f}".format(amin(vetorAcuracias)))
print("Máxima:\t\t", "{0:.5f}".format(amax(vetorAcuracias)))
print("Média:\t\t", "{0:.5f}".format(average(vetorAcuracias)))
print("Mediana:\t", "{0:.5f}".format(median(vetorAcuracias)))
print("Moda:\t\t", "{0:.5f}".format(stats.mode(vetorAcuracias)[0][0]))
print("Variância:\t", "{0:.5f}".format(var(vetorAcuracias)))
print("Desvio Padrão:\t", "{0:.5f}".format(std(vetorAcuracias)))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

params = {'figure.figsize': [10, 6], 
          'axes.labelsize': 16,
          'axes.titlesize':18, 
          'font.size': 16,
          'legend.fontsize': 12, 
          'xtick.labelsize': 12, 
          'ytick.labelsize': 12
         }

plt.rcParams.update(params)

#PLOTANDO A ACURÁCIA EM FUNÇÃO DA QTD DE NEURÔNIOS
plt.plot(range(1, 51), vetorAcuracias, label='Acurácia (Qtd Neurônios)')
plt.legend()
plt.title('Acurácia da Rede Neural')
plt.xlabel('Qtd de Neurônios')
plt.ylabel('Acurácia')
```
